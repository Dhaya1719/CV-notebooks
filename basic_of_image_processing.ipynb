```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import cv2
import numpy as np
import seaborn as sns
```

### WITHOUT MANUAL FUNCTION ###

##### LENNA ########

```python
image_bgr = cv2.imread('Lena.jpg')
image11 = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
grayscale_image11 = cv2.cvtColor(image11, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(9,9))
fig=plt.figure(1)
img1,img2=fig.add_subplot(121),fig.add_subplot(122)
img1.imshow(image11)
img2.imshow(grayscale_image11,plt.colormaps["grey"])
```

##### NATURE ########

```python
image_bgr = cv2.imread('NATURE.jpg')
image21 = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
grayscale_image21 = cv2.cvtColor(image21, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(9,9))
fig=plt.figure(1)
img1,img2=fig.add_subplot(121),fig.add_subplot(122)
img1.imshow(image21)
img2.imshow(grayscale_image21,plt.colormaps["grey"])
```

### WITH MANUAL FUNCTION ###

##### LENNA ########

```python
image_path = 'Lena.jpg'
image12 =imread(image_path)
r,g,b= image12[:,:,0],image12[:,:,1],image12[:,:,2]
gamma =1.5
r_const,g_const,b_const=0.2126,0.7152,0.0722
greyscale_image12 = r_const*r**gamma+g_const*g**gamma+b_const*b**gamma
```

```python
plt.figure(figsize=(9,9))
fig=plt.figure(1)
img1,img2=fig.add_subplot(121),fig.add_subplot(122)
img1.imshow(image12)
img2.imshow(greyscale_image12,plt.colormaps["grey"])
```

##### NATURE ########

```python
image22 =imread('NATURE.jpg')
r,g,b= image22[:,:,0],image22[:,:,1],image22[:,:,2]
gamma =1.5
r_const,g_const,b_const=0.2126,0.7152,0.0722
greyscale_image22 = r_const*r**gamma+g_const*g**gamma+b_const*b**gamma
```

```python
plt.figure(figsize=(19,9))
fig=plt.figure(1)
plt.figure(figsize=(19,9))
img1,img2=fig.add_subplot(121),fig.add_subplot(122)
img1.imshow(image22)
img2.imshow(greyscale_image22,plt.colormaps["grey"])
```

### HISTOGRAM EQUALIZATION ###

##### LENA ##### 

```python
histlena= cv2.equalizeHist(grayscale_image11)
original_gray_hist = cv2.calcHist([grayscale_image11], [0], None, [256], [0, 256])
equalized_gray_hist = cv2.calcHist([histlena], [0], None, [256], [0, 256])
plt.figure(figsize=(8,6))
plt.subplot(2, 2, 3)
plt.imshow(grayscale_image11, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(histlena, cmap='gray')
plt.title('Equalized Grayscale Image')
plt.axis('off')
plt.tight_layout()
plt.figure(figsize=(16,8))
plt.subplot(2, 2, 3)
sns.histplot(grayscale_image11.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.title('Histogram of Original Image')
plt.xlim([0, 256])
plt.subplot(2, 2, 4)
sns.histplot(equalized_gray_hist.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.title('Histogram of Equalized Image')
plt.xlim([0, 256])
plt.tight_layout()
plt.show()
```

```python
histogram, bins = np.histogram(grayscale_image11.flatten(), bins=256, range=[0, 256])
cdf = histogram.cumsum()
cdf_normalized = cdf * 255 / cdf[-1]  
equalized_image = np.interp(grayscale_image11.flatten(), bins[:-1], cdf_normalized).reshape(grayscale_image11.shape).astype(np.uint8)
plt.figure(figsize=(12, 6))
plt.figure(figsize=(8,6))
plt.subplot(2, 2, 3)
plt.imshow(grayscale_image11, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Grayscale Image')
plt.axis('off')
plt.tight_layout()
plt.figure(figsize=(16,8))
plt.subplot(2, 2, 3)
sns.histplot(grayscale_image11.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.title('Histogram of Original Image')
plt.xlim([0, 256])
plt.subplot(2, 2, 4)
sns.histplot(equalized_image.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.title('Histogram of Equalized Image')
plt.xlim([0, 256])
plt.tight_layout()
plt.show()
```
